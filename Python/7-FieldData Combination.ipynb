```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Set some directories. Here we use absolute directories. 
cwd = 'c:\\Users\\m1865\\Desktop\\Ticino'
cwd_Field = cwd + '\\FieldData'
cwd_Field_Data2 = cwd_Field + '\\Field Dataset 2'
cwd_Result = cwd + '\\Results'
```

# Process of Dataset 1

```python
df_Field1 = pd.read_excel(cwd_Field + '\\Sintesi_siti_ParcoTicino_con_specie.xlsx')
df_Field1.head()
```

```python
# Fill NaN with 0 to simplify the later process
df_Field1 = df_Field1.fillna(0)
df_Field1.head()
```

```python
# Convert the default float values to int64 values
df_Field1_Columns = df_Field1.columns
for column in df_Field1_Columns:
    if ("Piano Dominato" in column) or ("Piano Dominante" in column):
        if not ("Copertura" in column):
            df_Field1 = df_Field1.astype({column: 'int64'})
df_Field1.dtypes
```

```python
# Drop useless columns
for column in df_Field1_Columns:
    if not ((column == 'ESU') or ("Piano Dominato" in column) or ("Piano Dominante" in column) or (column == 'Lat') or (column == 'Long')):
        df_Field1.drop(columns = [column], inplace = True)
df_Field1
```

```python
# Get the rows where there are no data of the quantities of the species
df_Field1_Columns_Truncated = df_Field1.columns
list_Row_toDrop = []
for row in range(0,df_Field1.shape[0]):
    temp_Check_toDrop = True
    for column in df_Field1_Columns_Truncated:
        if ("Piano Dominato" in column) or ("Piano Dominante" in column):
            if not ("Copertura" in column):
                if df_Field1.loc[df_Field1.iloc[row].name,column] != 0:
                    temp_Check_toDrop = False
                    break
    if temp_Check_toDrop:
        list_Row_toDrop.append(row)
list_Row_toDrop
```

```python
# Drop those empty rows! 
df_Field1_Valid = df_Field1.drop(index = df_Field1.index[list_Row_toDrop])
df_Field1_Valid
```

```python
# Combine "Piano Dominante" and "Piano Dominato", merging the same species and keeping the unique species
list1_Spezie = []
for spezie in df_Field1_Valid.columns:
    if (not "Copertura" in spezie) and (" - " in spezie):
        list1_Spezie.append(spezie)
list1_Spezie
list_Spezie_Dominato = []
list_Spezie_Dominante = []
for spezie in list1_Spezie:
    if "Dominato" in spezie:
        list_Spezie_Dominato.append(spezie.split(" - ")[1])
    elif "Dominante" in spezie:
        list_Spezie_Dominante.append(spezie.split(" - ")[1])
# Sum up the counts of the same plants in "Piano Dominante" and "Piano Dominato". 
df_Field1_Merge = df_Field1_Valid.copy()
for spezie_Dominante in list_Spezie_Dominante:
    # Set a check value in case that a plant only exists in "Piano Dominante"
    spezie_unique = True
    for spezie_Dominato in list_Spezie_Dominato: 
        if spezie_Dominante == spezie_Dominato:
            # Add up the counts if the plant exists in both "Piano Dominante" and "Piano Dominato"
            df_Field1_Merge[spezie_Dominante] = df_Field1_Merge[("Piano Dominante - " + spezie_Dominante)] + df_Field1_Merge[("Piano Dominato - " + spezie_Dominato)]
            # Drop the two columns! 
            df_Field1_Merge = df_Field1_Merge.drop(columns = [("Piano Dominante - " + spezie_Dominante),("Piano Dominato - " + spezie_Dominato)])
            spezie_unique = False
            break
    # If the plant only exists in "Piano Dominante", then no need to add any arithmetric operation! 
    if spezie_unique:
        df_Field1_Merge[spezie_Dominante] = df_Field1_Merge[("Piano Dominante - " + spezie_Dominante)]
        df_Field1_Merge = df_Field1_Merge.drop(columns = [("Piano Dominante - " + spezie_Dominante)])
# Now only the plants unique in "Piano Dominato" remain intact in the geodataframe. We simply rename their names! 
for column in df_Field1_Merge.columns: 
    if ("Piano Dominato" in column) and (not "Copertura" in column):
        df_Field1_Merge = df_Field1_Merge.rename(columns={column: column.split(" - ")[1]})
# Finally check our columns of the final dataframe! 
df_Field1_Merge
```

```python
# Convert the species columns into rows, so that the new dataframe will only have 5 columns, where the fourth column will be the name of the species and the fifth column will be the quantity of that species
list1_FOR = []
list1_Lat = []
list1_Long = []
list1_species = []
list1_quantity = []
for row in range(0,df_Field1_Merge.shape[0]):
    for column in df_Field1_Merge.columns:
        if not (column == 'ESU') and not (column == 'Lat') and not (column == 'Long'):
            if not ("Copertura" in column) and (df_Field1_Merge.loc[df_Field1_Merge.iloc[row].name,column] != 0):
                list1_FOR.append(df_Field1_Merge.iloc[row,0])
                list1_Lat.append(df_Field1_Merge.iloc[row,1])
                list1_Long.append(df_Field1_Merge.iloc[row,2])
                list1_species.append(column)
                list1_quantity.append(df_Field1_Merge.loc[df_Field1_Merge.iloc[row].name,column])
df_Field1_Converted = pd.DataFrame(
    data = {
        'Plot': list1_FOR,
        'Lat': list1_Lat,
        'Long': list1_Long,
        'Species': list1_species,
        'Quantity': list1_quantity
    }
)
df_Field1_Converted
```

```python
# Convert the latitude and longitude to the correct form
df_Field1_Converted['Lat'] = df_Field1_Converted['Lat'].str.split(r'°|\'', regex = True, expand = True).iloc[:,0].apply(pd.to_numeric) + df_Field1_Converted['Lat'].str.split(r'°|\'', regex = True, expand = True).iloc[:,1].apply(pd.to_numeric).apply(lambda row: row/60)
df_Field1_Converted['Long'] = df_Field1_Converted['Long'].str.split(r'°|\'', regex = True, expand = True).iloc[:,0].apply(pd.to_numeric) + df_Field1_Converted['Long'].str.split(r'°|\'', regex = True, expand = True).iloc[:,1].apply(pd.to_numeric).apply(lambda row: row/60)
df_Field1_Converted
```

# Process of Dataset 2

```python
# Since our Excel file contains multiple sheets, we need to get the names of sheets firstly. 
xlsx = pd.ExcelFile(cwd_Field_Data2 + '\\Dati plot forestali Parco Ticino completo.xlsx')
xlsx.sheet_names
```

```python
# Combine! 
xlsx_field = pd.DataFrame()
for sheet in xlsx.sheet_names:
    if sheet != 'TEMPLATE':
        temp_df = pd.read_excel(cwd_Field_Data2 + '\\Dati plot forestali Parco Ticino completo.xlsx', sheet_name=sheet)
        xlsx_field = pd.concat([xlsx_field, temp_df], ignore_index=True)
xlsx_field
```

```python
# By observing our concatenated dataframe, we only need to keep the first 3 columns. 
# Drop those columns! 
xlsx_field = xlsx_field.drop(xlsx_field.iloc[:,np.r_[1,np.arange(3,6,1)]], axis = 1)
xlsx_field
```

```python
# Now we need to drop the useless rows. If the first column of a row is not a numeric value, then it will be dropped. 
df_Field2 = xlsx_field.copy()
for row in range(0,xlsx_field.shape[0]):
    # Get the index label of the current row! 
    temp_label = xlsx_field.iloc[row].name
    if str(xlsx_field.iloc[row,0]).isnumeric(): # Note: we have to convert the value to str first, otherwise we can't apply "isnumeric()" function. 
        continue
    else: 
        # Drop the row based on the label! 
        df_Field2 = df_Field2.drop(index=temp_label, axis = 0)
# Reset index
df_Field2 = df_Field2.reset_index(drop = True)
# Assign column names
df_Field2.columns = ["Plot Number", "Tree Species"]
# Finally we get our clean dataframe! 
df_Field2
```

```python
# Now we group by to aggregate some values
df_Field2 = df_Field2.groupby(['Plot Number', 'Tree Species'], as_index = False).size()
df_Field2 = df_Field2.rename(columns = {'size': 'Species Counts'})
df_Field2
```

```python
# Read another Excel file which contains all the geographical coordinates of the field plots! 
df_field_coord = pd.read_excel(cwd_Field_Data2 + "\\Dati coordinate plot forestali Parco Ticino.xlsx")
df_field_coord
```

```python
# Rename the columns !
df_field_coord.columns = ["Plot Number", "Latitude", "Longitude"]
```

```python
# Merge! 
df_field_merged = pd.merge(df_Field2, df_field_coord, on = "Plot Number")
df_field_merged
```

```python
# Create immediately the geodataframe! Since we need to filter out the field plots which are ouside our RoI! 
gdf_Merged = gpd.GeoDataFrame(
    df_field_merged,
    geometry = gpd.points_from_xy(df_field_merged['Longitude'],df_field_merged['Latitude']),
    crs = 'EPSG:4326'
)
# Convert the crs from 4326 (WGS84) to 32632 (WGS84 UTM zone 32N)
gdf_Merged = gdf_Merged.to_crs(32632)
gdf_Merged
```

```python
# Load our region of interest shapefile! 
gdf_RoI = gpd.read_file(cwd_Field + "\\ShapefileCorretti\\confini_foreste_corretti.shp")
gdf_RoI
```

```python
# Check which field plots (points) are within our RoI! 
join_inner_df = gdf_RoI.sjoin(gdf_Merged, how="inner").sort_values('Plot Number').groupby('Plot Number', as_index = False).count()
list_ValidPoints = list(join_inner_df['Plot Number'])
list_ValidPoints
```

```python
# Back to our field dataframe and remove those rows which are outside our RoI! 
df_field_Valid = df_field_merged.copy()
df_field_Valid = df_field_Valid.where(df_field_Valid['Plot Number'].isin(list_ValidPoints)).dropna().reset_index(drop = True)
# Don't know why the datatype is changed to float from int. Now we change back to int! 
df_field_Valid['Plot Number'] = df_field_Valid['Plot Number'].astype('int64')
df_field_Valid['Species Counts'] = df_field_Valid['Species Counts'].astype('int64')
# Finally we have our field plot dataframe with only valid points! 
df_field_Valid
```

```python
df_Field2_Final = df_field_Valid.copy()
df_Field2_Final = df_Field2_Final.rename(columns={'Plot Number': 'Plot', 'Tree Species': 'Species', 'Species Counts': 'Quantity', 'Latitude': 'Lat', 'Longitude': 'Long'})
df_Field2_Final
```

# Concatenate 2 datasets

```python
df_Field = pd.concat([df_Field1_Converted, df_Field2_Final])
df_Field
```

# Create geodataframe, add buffer (5 pixels = 50m diameters) and export to shapefile

```python
df_Field_Unique = df_Field.drop(columns = ['Species','Quantity']).drop_duplicates().reset_index()
```

```python
gdf_Field = gpd.GeoDataFrame(
    df_Field_Unique,
    geometry = gpd.points_from_xy(df_Field_Unique['Long'],df_Field_Unique['Lat']),
    crs = 'EPSG:4326'
)
gdf_Field
```

```python
gdf_Buffer = gdf_Field.copy().to_crs(32632)
gdf_Buffer['geometry'] = gdf_Buffer.geometry.buffer(25, cap_style = 3)
gdf_Buffer.to_file(cwd_Field + '\\Buffer.shp')
gdf_Buffer
```
