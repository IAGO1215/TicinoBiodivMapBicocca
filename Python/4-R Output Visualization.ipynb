```python
import os
import matplotlib.pyplot as plt
import geopandas as gpd
from osgeo import gdal
import rasterio as rio
from rasterio.plot import show
import rasterio.mask
```

```python
cwd = 'c:\\Users\\m1865\\Desktop\\Ticino'
cwd_Field = cwd + '\\FieldData'
cwd_Field_Correct = cwd_Field + '\\ShapefileCorretti'
cwd_ProcessedRaster = cwd + '\\ProcessedData'
cwd_Result = cwd + '\\Results'
```

```python
shp_Field = gpd.read_file(cwd_Field_Correct + "\\confini_foreste_corretti.shp")
```

```python
name_raster = os.listdir(cwd_Result)
name_raster
```

```python
raster_Spectral = []
for x in name_raster:
    temp = rio.open(cwd_Result + "\\" + x + "\\SPCA\\SpectralSpecies\\SpectralSpecies")
    fig, ax = plt.subplots(1, figsize = (10, 10))
    fig.suptitle(x + " Spectral Species")
    fig.tight_layout()
    show(temp, ax = ax, cmap = 'turbo')
    raster_Spectral.append(temp)
```

```python
raster_Alpha = []
for x in name_raster:
    temp = rio.open(cwd_Result + "\\" + x + "\\SPCA\\Alpha\\Shannon_10")
    fig, ax = plt.subplots(1, figsize = (10, 10))
    image_hidden = ax.imshow(temp.read()[0], 
                         cmap='turbo')
    show(temp, ax = ax, cmap = 'turbo')
    fig.suptitle(x + " Shannon Index")
    fig.colorbar(image_hidden, ax=ax, fraction=0.046, pad=0.04)
    fig.tight_layout()
    raster_Alpha.append(temp)
```

```python
raster_Beta = []
for x in name_raster:
    temp = rio.open(cwd_Result + "\\" + x + "\\SPCA\\Beta\\BetaDiversity_BCdiss_PCO_10")
    fig, ax = plt.subplots(1, figsize = (10, 10))
    image_hidden = ax.imshow(temp.read()[0], 
                         cmap='turbo')
    show(temp, ax = ax, cmap = 'turbo')
    fig.suptitle(x + " Beta Diversity Bray-Curtis Dissimilarity")
    fig.colorbar(image_hidden, ax=ax, fraction=0.046, pad=0.04)
    fig.tight_layout()
    raster_Beta.append(temp)
```
